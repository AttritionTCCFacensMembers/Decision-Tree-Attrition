#### 2.2 - Tratando datasets in_time e out_time para gerar variáveis relevantes

Conforme pesquisado na literatura, workload é um dos principais fatores de attrition. Portanto, através dos datasets de in_time e out_time, será criado um atributo para avaliar o workload dos funcionários.

Com as informações de horário de entrada e saída de cada um dos funcionário, vamos avaliar primeiramente qual é o tempo médio de trabalho de todos os funcionários da companhia e em seguida, avaliar qual é o tempo médio de horas trabalhadas a mais ou a menos por dia (em relação à media geral), de cada um dos funcionários. 

##### 2.2.1 - tratativa de in_time e out_time fase 1/3

```python
# tratativa de in_time e out_time fase 1/3
#Carregando datasets de ponto:
df_in = pd.read_csv('in_time.csv')
df_out= pd.read_csv('out_time.csv')
#Renomeando colunas de ID para facilitar:
df_in.rename(columns = {'Unnamed: 0': 'EmployeeID'},inplace= True)
df_out.rename(columns = {'Unnamed: 0': 'EmployeeID'}, inplace = True)

#Transformando todas as colunas que representam os dias  em datetime (menos ID):
cols_in = df_in.columns[1:]
cols_out = df_out.columns[1:]

for i in cols_in:
    df_in[i] = pd.to_datetime(df_in[i])
    
for i in cols_out:
    df_out[i] = pd.to_datetime(df_out[i])

#valida número de colunas
print(' Colunas hora de entrada:',len(cols_in),'\n','Colunas hora de Saída:',len(cols_out))

#Dataset com todos os tempos para fazer operações
df_time = df_out.merge(df_in,on ='EmployeeID', how='left', suffixes=('_out','_in'))

#criando colunas com tempo trabalhado por dia e deletando In_time e Out_time
for i in cols_out:
    df_time[i] = df_time[i+'_out'] - df_time[i+'_in']
    df_time.drop([i+'_out',i+'_in'],axis=1,inplace=True)

#A primeira fase da tratativa nos trouxe um dataset com o tempo trabalhado por dia de cada funcionário
#Colunas representando o dia e linhas os funcionários
df_time.head()
```

##### 2.2.2 - tratativa de in_time e out_time fase 2/3

```python
# # tratativa de in_time e out_time fase 1/3
# a segunda fase da tratativa visa fazer o agrupamento dos dias em meses e ano
#gerando variáveis de dias em forma de datetime
t_cols = pd.to_datetime(df_time.columns[1:])
#Identificando o mes de cada coluna
month = list(t_cols.month)
cols = df_time.columns[1:]

#Gerando colunas da média de tempo e números de dias trabalhados no ano
df_time['avg_wrk_time_year'] = df_time[cols].mean(axis=1)
df_time['worked_days_year'] = df_time[cols].count(axis=1)

# loop nos 12 meses do ano gerando as mesmas colunas acima por mês (média de tempo e número de dias)
for mes in range(1,13):
#achar o indice das colunas de cada dia dentro do loop mês para filtrar e guardar em uma lista
    month_idx = [i for i, e in enumerate(month) if e == mes]
    
    #o primeiro indice de coluna do df_time é o Employee id, então vamos add 1 em cada indice da month_idx para offsetar
    for idx,i in enumerate(month_idx):
        month_idx[idx] = i+1
        # criando média de tempo trabalhado no mês (foi utilizado o loc nas colunas e a média no eixo de coluna)
        df_time['avg_wrk_time_'+str(mes)] = df_time.iloc[:,month_idx].mean(axis=1)
        #criando total de dias trabalhados no mês (mesmo processo da linha acima porém invés de média utilizamos o count)
        df_time['work_days_'+str(mes)] = df_time.iloc[:,month_idx].count(axis=1)

#O resultando é o df_time com novas colunas de média de tempo trabalhos e dias por mês e ano
df_time.head()
```

##### 2.2.3 - tratativa de in_time e out_time fase 3/3

```python
from datetime import date
#importando feriados de 2015
#Para metodos de como usar a biblioteca holidays e entendimento do seu funcionamento utilize o link
#https://github.com/dr-prodigy/python-holidays
import holidays
# Selecionando país e ano que queremos descobrir os feriados (alinhado ao nosso dataset)
us_holidays = holidays.US(years=2015)

#utilizando as colunas de dias criadas na fase 2/3 para gerar uma lista com os feriados do ano
t_holiday = []
for i in t_cols:
    #print('date:',i,i in us_holidays)
    if i in us_holidays:
        t_holiday.append(i)
for i in t_holiday:
    #dando o print dos feriados localizados
    print('date:',i,'Holiday:',us_holidays.get(i))

    
#transformando timestamp da lista de feriados em texto    
cols_holiday = []
for i in range(len(t_holiday)):
    #print(str(t_holiday[i])[:-9])
    cols_holiday.append(str(t_holiday[i])[:-9])

#criando coluna de feriados trabalhados e média de tempo trabalhada em feriado por funcionário no ano  
df_time['holidays_worked'] = df_time[cols_holiday].count(axis=1)
df_time['avg_holidays_work_time'] = df_time[cols_holiday].mean(axis=1)

#dando o drop das colunas de dias do df_time utilizando as colunas guardadas na fase 2/3 da tratativa
df_time.drop(cols,axis=1,inplace=True)

#criando uma coluna da diferença de trabalho por funcionário vs a média geral de trabalho do total de empregados da empresa (no ano)
df_time['delta_min_year'] = df_time['avg_wrk_time_year']-df_time['avg_wrk_time_year'].mean()

#transformando a coluna do delta de tempos em minutos
for i,i2 in enumerate(df_time['delta_min_year']):
    df_time['delta_min_year'][i] = i2.total_seconds()/60
    
#transformando a coluna delta em float    
df_time['delta_min_year'] = df_time['delta_min_year'].astype(float)

#com isso transformamos o df_time de 288 colunas em 30 colunas que são as seguintes

#EmployeeID: ID do funcionário para dar merge com o dataset principal
#avg_wrk_time_year: média de tempo trabalhado por dia no ano
#worked_days_year: número de dias trabalhados no ano
#holidays_worked: número de feriados trabalhados no ano
#avg_holidays_work_time: média de tempo trabalhado por dia nos feriados do ano
#delta_min_year: diferença de tempo no ano que esse colaborador tem entre a média trabalhada por dia no ano de toda empresa (em minutos)
#avg_wrk_time_n: média de tempo trabalhado por dia no mês (sendo n o numero do mês)
#work_days_n: número de dias trabalhados no mês (sendo n o número do mês)

df_time.head()
```
